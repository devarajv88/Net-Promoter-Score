# **Airbnb- Net Promoter Score Calculation**



# **Project Description**
* Given dataset contains the feedback received by the customers of airbnb and we are supposed to calculate the Net Promoter Score (NPS) for the business based on the feedbacks received.



# **Approach**
* For calulation of NPS, any rating of less than 7 will be considered as detractors.
* Rating of 7 or 8 will be categorized under passive category
* And ratings of 9 or 10 will be considered as promoters
* Net Promoter Score (NPS) = percentage of promoters - percentage of detractors.



```python
#Importing relevent libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing Dataset
data = np.loadtxt("/kaggle/input/airbnb-feedback/survey.txt")
data
```

```python
print(f"Maximum value is: {np.max(data)}")
print(f"Minimum value is: {np.min(data)}")
print(f"Mean value is: {np.round(np.mean(data))}")
```

```python
# Get the unique values and their frequencies
unique_values, frequencies = np.unique(data, return_counts=True)

# Plot the histogram
plt.bar(unique_values, frequencies, edgecolor='black')
plt.xticks(unique_values)
plt.xlabel('Ratings')
plt.ylabel('Frequencies')
plt.title('Feedback Ratings Frequencies')
plt.grid(axis='y')
plt.show()
```

```python
total = data.shape[0]
print(f"Total number of people participated in the survey: {total}")
```

```python
promoter_percentage = np.round(data[data > 8].shape[0] * 100 / total, 2)
print(f"Percentage of promoters: {promoter_percentage}")
```

```python
detractor_percentage = np.round(data[data < 7].shape[0] * 100 / total, 2)
print(f"Percentage of detracters: {detractor_percentage}")
```

```python
passive_percentage = np.round(data[(data > 6) & (data < 9)].shape[0] * 100 / total, 2)
print(f"Percentage of passives: {passive_percentage}")
```

```python
#NPS = promoter% - detractor%
nps = np.round((promoter_percentage - detractor_percentage), 2)
print(f"Based on the {total} number of customer feedbacks recevied, NPS for Air bnb is found to be {nps}.")
```

- **Based on the given dataset of 1167 records, NPS of Airbnb is calculated to be 23.74**

- **For better understanding, we could compare this NPS score with the previous month's scores to understand about the services provided and the customer satisfaction. Also comparing this NPA value with that of competitors from the same sector, could show some more insights.**
